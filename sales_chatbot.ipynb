```python
allData = []

def add_data1(data, productType):
  for val in data:
     x = "Product Type : {}\n".format(productType)
     for key, value in val.items():
        if type(value) != type(""):
          value = ' . '.join(value)
        x = x + "{} : {}\n".format(key, value)
     allData.append(x)

def add_data2(data, productType):
  for val in data['products']:
    x = "Product Type : {}\n".format(productType)
    for key, value in val.items():
      x = x + "{} : {}\n".format(key, value)
    allData.append(x)

add_data1(body_data, "BODY")
add_data2(hands_feet_data, "HANDS AND FEET")
add_data2(underarm_data, "UNDERARMS")
```

```python
index = pc.Index('llama-2-rag')

embeds = embed_model.embed_documents(allData)
ids = [str(i) for i in range(len(allData))]
# get metadata to store in Pinecone
metadata = [
    {'text': allData[i]} for i in range(len(allData))
]
# add to Pinecone
index.upsert(vectors=zip(ids, embeds, metadata))
```

```python
def get_context(query):
  query_embed = embed_model.embed_documents(query)[0]
  response = index.query(
    vector=query_embed,
    top_k=3,
    include_values=False,
    include_metadata=True
  )["matches"]
  results = [x['metadata']['text'] for x in response]
  context = '\n\n'.join(results)
  return context
```

```python
from typing import List, Dict, Any

class ExtendedConversationBufferWindowMemory(ConversationBufferWindowMemory):
    extra_variables: List[str] = []

    @property
    def memory_variables(self) -> List[str]:
        """Will always return list of memory variables."""
        return [self.memory_key] + self.extra_variables

    def load_memory_variables(self, inputs: Dict[str, Any]) -> Dict[str, Any]:
        """Return buffer with history and extra variables"""
        d = super().load_memory_variables(inputs)
        d.update({k:inputs.get(k) for k in self.extra_variables})
        return d
```
